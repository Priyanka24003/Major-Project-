```python
# check for categorical attributes
cat_col = []
for x in df.dtypes.index:
    if df.dtypes[x] == 'object':
        cat_col.append(x)
cat_col







# name class section 
```

```python
# fill the missing values
item_weight_mean = df.pivot_table(values = "Item_Weight", index = 'Item_Identifier')
item_weight_mean
```

```python
miss_bool = df['Item_Weight'].isnull().value_counts()
miss_bool
```

```python
for i, item in enumerate(df['Item_Identifier']):
    # Ensure the index exists in the miss_bool array
    if i < len(miss_bool) and miss_bool[i]:  # Check if index exists and if condition is true
        if item in item_weight_mean:
            df.loc[i, 'Item_Weight'] = item_weight_mean.loc[item, 'Item_Weight']
        else:
            df.loc[i, 'Item_Weight'] = np.mean(df['Item_Weight'])
```

```python
df['Item_Weight'].isnull().sum()
```

```python
outlet_size_mode = df.pivot_table(values='Outlet_Size', columns='Outlet_Type', aggfunc=(lambda x: x.mode()[0]))
outlet_size_mode
```

```python
miss_bool = df['Outlet_Size'].isnull()
df.loc[miss_bool, 'Outlet_Size'] = df.loc[miss_bool, 'Outlet_Type'].apply(lambda x: outlet_size_mode[x])
```

```python
df['Outlet_Size'].isnull().sum()
```

```python
sum(df['Item_Visibility']==0)
```

```python
# replace zeros with mean
df.loc[:, 'Item_Visibility'].replace([0], [df['Item_Visibility'].mean()], inplace=True)
```

```python
sum(df['Item_Visibility']==0)
```

```python
# combine item fat content
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace({'LF':'Low Fat', 'reg':'Regular', 'low fat':'Low Fat'})
df['Item_Fat_Content'].value_counts()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
train = pd.read_csv('train.csv')

# Define categorical columns
categorical_cols_to_display = train.select_dtypes(include=['object', 'category']).columns.tolist()

# Plotting pie charts without the first plot and without black edges
fig, ax = plt.subplots(2, 2, figsize=(12, 12))  # 2x2 grid (4 plots total)
new_palette = ['gold', 'skyblue', 'coral', 'limegreen']

for i, col in enumerate(categorical_cols_to_display[1:5]):  # Skip the first column
    train[col].value_counts().plot.pie(
        autopct="%.2f%%",
        ax=ax[i//2, i%2],
        colors=new_palette,
        wedgeprops={'edgecolor': 'none'}  # No black edges
    )
    ax[i//2, i%2].set_title(f'{col} Distribution', fontsize=20)

plt.tight_layout()
plt.show()
```
